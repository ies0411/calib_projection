```python
import math
import quaternion
#conda install quaternion
#conda install -c conda-forge quaternion
#python -m pip install --user --upgrade --force-reinstall numpy-quaternion
```

```python

for idx,result in enumerate(result_list[0]):
  object_dic = copy.deepcopy(object_template)
  object_dic['tracking_id'] = idx
  object_dic["id"] = str(uuid.uuid4())
  object_dic['frames'][0]['annotation']['coord']['position']['x'] = result[3]
  object_dic['frames'][0]['annotation']['coord']['position']['y'] = result[4]
  object_dic['frames'][0]['annotation']['coord']['position']['z'] = result[5]
  object_dic['frames'][0]['annotation']['coord']['size']['x'] = result[0]
  object_dic['frames'][0]['annotation']['coord']['size']['y'] = result[1]
  object_dic['frames'][0]['annotation']['coord']['size']['z'] = result[2]

  theta = result[6]
  euler_obj = [0,0,theta]
  quat_obj = quaternion.from_euler_angles(euler_obj)
  object_dic['frames'][0]['annotation']['coord']['rotation_quaternion']['x'] = quat_obj.x
  object_dic['frames'][0]['annotation']['coord']['rotation_quaternion']['y'] = quat_obj.y
  object_dic['frames'][0]['annotation']['coord']['rotation_quaternion']['z'] = quat_obj.z
  object_dic['frames'][0]['annotation']['coord']['rotation_quaternion']['w'] = quat_obj.w


  label_info_to_upload['result']['objects'].append(object_dic)
  




```

```python
class_to_class_id = {
    class_name: project_id,
}
class_to_class_color = {
    class_name: "#FF625A",
}


classes_id = set()
classes = set()
classes_count = {}
for obj in label_info_to_upload["result"]["objects"]:
    classes_id.add(obj["class_id"])
    classes.add(obj["class_name"])
if obj['class_name'] in classes_count: 
    classes_count[obj["class_name"]] += 1
else:
    classes_count[obj["class_name"]] = 1
label_info_to_upload["tags"]["classes_id"] = list(classes_id)
label_info_to_upload["tags"]["class"] = list(classes)
label_info_to_upload["tags"]["classes_count"] = [{ "id": class_to_class_id[k], "name": k, "count": v } for k, v in classes_count.items()]        
```
